# 2-NUMBER OF EMPLOYEES DATA: importing and cleanning data

```python
def f_employ(x): # function to import and clean data
    df=pd.read_excel(x,sheet_name='employés')
    #date=[];ddates=[]
    def col_names(df): # look for the number of line with the names of the banks to be columns names
        y=[];z=[];
        for i in range(0,(df.shape[0]-1)):
            if 'BNC' in df.loc[i:(i+1),:].values:
                y=df.loc[i:(i+1),:].values.tolist()[0]
                z=i+1
                #break
        # add Date as name of column with the dates
        df.columns=['Date']+y[1:]#df_employ.columns[1:].tolist()
        df=df.iloc[z:,:]
        return df
    df=col_names(df)

    # remove columns where names are NaN
    jj=[ type(i)!=float for i in df.columns.values]
    df=df[df.columns.values[jj]]
    # remove all rows where there is at least 10 NaN
    df=df.dropna(axis=0,thresh=10)
    
    date=[];ddates=[]
    def date_identify():
        ddates=df.Date.tolist()
        Sdate=pd.Series([3,6,9,12],index=['Mar','Jui','Sep','Déc'])
        month=[];year=[];days_length=[]
        for i in ddates:#dd.Date.tolist():
            month.append(Sdate[i[0:3]])
            year.append(int(i[-4:]))

        days_length=[];date=[]
        for j in range(0,len(month)):
            days_length.append(calendar.monthrange(year[j],month[j])[1])
            date.append(datetime(year[j],month[j],days_length[j]).strftime('%Y-%m-%d'))

        return date
    df.Date=date_identify()
    
    return df
```

```python
def f_employ2(x): # function to import and clean data
    df=pd.read_excel(x,sheet_name='employés')
    #date=[];ddates=[]
    def col_names(df): # look for the number of line with the names of the banks to be columns names
        y=[];z=[];
        for i in range(0,(df.shape[0]-1)):
            if 'BNC' in df.loc[i:(i+1),:].values:
                y=df.loc[i:(i+1),:].values.tolist()[0]
                z=i+1
                #break
        # add Date as name of column with the dates
        df.columns=['Date']+y[1:]#df_employ.columns[1:].tolist()
        df=df.iloc[z:,:]
        return df
    df=col_names(df)

    # remove columns where names are NaN
    jj=[ type(i)!=float for i in df.columns.values]
    df=df[df.columns.values[jj]]
    # remove all rows where there is at least 10 NaN
    df=df.dropna(axis=0,thresh=10)
    ###############
    date=[];ddates=[]
    def date_identify():
        ddates=df.Date.tolist()
        Sdate=pd.Series([3,6,9,12],index=['Mar','Jui','Sep','Déc'])
        month_first=[];year_first=[];month_end=[];year_end=[]
        for i in range(0,len(Sdate)):             # find the number of the months for first row of data
            if Sdate.index[i]==ddates[0][0:3]: 
                month_first.append(Sdate[i])
                year_first.append(ddates[0][-4:])
            if Sdate.index[i]==ddates[len(ddates)-1][0:3]:
                month_end.append(Sdate[i])
                year_end.append(ddates[len(ddates)-1][-4:])
        date_first=year_first[0]+'-'+str(month_first[0])+'-'+'01'    
        date_end=year_end[0]+'-'+str(month_end[0])+'-'+'01'
        #date_end=date_end.tolist()
        #date_end=date_end[len(date_end)-1]+relativedelta(months=3)
        period=pd.date_range(date_first,date_end,freq='Q').tolist()
        period.append(period[len(period)-1]+relativedelta(months=3))
        return period
    df.Date=date_identify()
    return df#ate_end#[0]+relativedelta(months=1)
```

### Arranging cleanning data in three columns: Date,Banks, Indicator

```python
def group_data(x,fun): # function to group data of a variable by bank
    df=fun(x)          # x is a data frame of and indicator, where the dates are in the first column and each bank is a column of the indicator
    dfs=df.columns[1:] # list of columns, except the dates
    date=[];values=[];banks=[]
    for d in range(0,len(dfs)):
        date.append(df.Date)
        values.append((df[dfs[d]]))
        banks.append([dfs[d]]*df.shape[0])
    def flatten_list(x): # function to flatten a list of lists to 
        y=[]
        for i in x:
            for j in i:
                y.append(j)
        return y
    date=flatten_list(date)
    values=flatten_list(values)
    banks=flatten_list(banks)
    df=pd.DataFrame({'Date':date,'Banks':banks,'Indicator':values})
    return df
```

### A Function to import and combine data

In this section all data related to the number of employees are imported and combined in one data frame named *df_employees*.

```python
def f_combine(x,fun): # x is the list of files names; fun is the function that import and clean data
    df=pd.DataFrame(columns=['Date','Banks','Indicator'])
    for file in x:
        df=pd.merge(df,group_data(file,fun),how='outer')
    return df
```

### Function to combine the available data with new quarterly report

```python
def add_employ(newReportFile):
    df_new=group_data(newReportFile,f_employ3)      # importing and cleanning data
    #df_new1=df_new.drop_duplicates(subset=['Date','Banks']) # droping duplicated data, by Date and Banks
    df_emp=pd.read_csv('BanksEmployees.csv')        # importing old data before combining with new data
    df=pd.concat([df_emp,df_new],join='inner')
    df.Date=df.Date.tolist()
    df=df.drop_duplicates(subset=['Date','Banks']) # droping duplicated data, by Date and Banks
    df.to_csv('BanksEmployees.csv',date_format='%Y-%m-%d',index=False)
    return df
```

# 3- BALANCE SHEET DATA: importing and cleanning data

```python
def f_balsheet(x): # function to import and clean data
    
    # looking for the right sheet name of balance sheet for the quarterly which is the name of the Excel file
    df=pd.ExcelFile(x)
    sheets=df.sheet_names
    def f_removeSpacesInList(x): # funtion to remove spaces in each string of a list
        y=[]
        for i in x:
            y.append(i.replace(' ',''))
        return y
    sheets1=f_removeSpacesInList(sheets)   # removing spaces in each sheet name to identify the position of the sheet name needed easily
    # for there is a space between the month abbreviation and the two last numbers of the years in some reports, and this space is not present in other reports.
    Strim=pd.Series([1,1,2,2,3,4,4],index=['dec.','déc.','mars','nars','juin','sept.','sept']) # serie to build report name
    # looking for the right sheet name of the balance sheet for the quarterly
    trim=int(x.split('.')[0][5])
    year=x.split('.')[0][-2:]
    year_quarter=pd.Series([year[-2:],year[-2:],year[-2:],str(int(year)-1)[-2:]],index=[2,3,4,1]) # to identify the year, 
    # remembering that when quarter= 1 for year 2015 is the quarter ending on December of the last civil year, then December 2014
    year_true=year_quarter[trim]
    if len(year_true)==1:
        year_true='0'+year_true # for years 2001, 2002 and so on, year_true is a string of the last number. This line of code correct it to have zero and the last number of the year.
    z=[] # last month of quarterly - The first quarterly start on October and end in December, according to the fiscal year in Haiti
    for i in range(0,len(Strim)):
        if Strim[i]==trim:
            z.append(Strim.index[i])
    
    psn=[] # possible sheet name(s) of the balance sheet 
    for i in z:
        psn.append('bilsysban'+i+year_true)
        psn.append('bilsysba'+i+year_true)  # at least in one report the correct sheet name structure is different
    # looks for the name of of psn which is in sheet1 and use the same position index of this name in sheet1 to retrieve the right name in the list of sheets
    # looks for the name of of psn which is in sheet1 and use the same position index of this name in sheet1 to retrieve the right name in the list of sheets
    sheetName=[]
    for position in range(0,len(sheets1)):
        for sh_name in psn:
            if sh_name==sheets1[position]:
                sheetName.append(sheets[position])
    df=pd.read_excel(x,sheet_name=sheetName[0])
    
      # stop the data frame at row where is found the string "Dépôts US convertis".
    df=df.reset_index()
    endrow_data=[]
    for i in range(0,df.shape[0]):
        if 'Dépôts US convertis' in df.loc[i:i,:].values.tolist()[0]:
            endrow_data=i
    df=df.loc[:endrow_data,:] 
    
    # Give names to columns
    def startswith_InList(x,starts): # to verify if there is at least an element of the list of 
    # strings x, starting with the string indicated by starts
        y=[]
        for i in range(0,len(x)):
            if type(x[i])==str and x[i].startswith(starts):
                y=True
                break
        if y==True:
            return True
        else:
            return False
    
    def col_names(df): # look for the number of line with the names of the banks to be columns names
            #y=[];
            z=[]
            for i in range(0,(df.shape[0]-1)):
                #if 'BNC  ' in df.loc[i:(i+1),:].values.tolist()[0]:
                if startswith_InList(df.loc[i:i,:].values.tolist()[0],'BNC')==True:
                    y=df.loc[i:i,:].values.tolist()[0]
                    z=i+1    # defining row to start data frame
                    #break
            # add Date as name of column with the dates
            df.columns=['Index','Indicators']+y[2:]#df_employ.columns[1:].tolist()
            df=df.iloc[z:,1:] # begin data frame with row z and column of indicators names
            return df
    
    df=col_names(df)
      
    # remove columns where names are NaN
    jj=[ type(i)!=float for i in df.columns.values]
    df=df[df.columns.values[jj]]
    ####
     #####
    def lower_removeAccentSpacelist(x):# function to lower case and remove accents 
        # and spaces in indicators names - x is a list
        y=[]
        for i in range(0,len(x)):
            y.append(unidecode(str(x[i])).lower().strip())
        return y
    df.Indicators=lower_removeAccentSpacelist(df.Indicators.tolist())
    
    ### Create a column indicating bilance sheets parts
    def balsheet_parts(x,parts1):
    
        def f_grouped1(x,start,until,fill): # x is a list - start is a string positio of the list x where to start repeating the string argument named 'fill', and stop until before the position index of the string argument 'until'.
                h=[]
                for i in range(0,len(x)):
                    if x[i]==start: #lower_removeAccentSpace(i)!=until:
                        for j in range(i,len(x)):
                            if x[j]!=until:
                                h.append(fill)
                            if x[j]==until:
                                break
                return h
        y=[]
        for j in range(0,len(parts1)):
                if (j+1)<len(parts1):
                    y.append(f_grouped1(x=x,start=parts1[j],until=parts1[j+1],fill=parts1[j])) # filled with the same string than the argument 'start'
                elif (j+1)==len(parts1):
                    y.append(f_grouped1(x=x,start=parts1[j],until='',fill=parts1[j]))

        z=[]
        for i in range(0,len(y)):
            for j in range(0,len(y[i])):
                z.append(y[i][j]) 
        return z
    df['Parts1_fr']=balsheet_parts(df.Indicators.tolist(),parts1=['actif','passif','avoir des actionnaires','hors-bilan',
                          'portefeuille brut','depots us convertis'])

    df['Parts2_fr']=balsheet_parts(df.Indicators.tolist(),parts1=['actif','disponibilites',
                              'bons b.r.h.','depots', 'obligations a vue','obligations a terme',
 'autres passifs','avoir des actionnaires','hors-bilan','portefeuille brut',
 'depots us convertis'])
#################################
    def fparts1_en(): # translation in English
        y=[]
        for i in df.Parts1_fr.tolist():
            if i=='actif':
                y.append('Asset')
            elif i=='passif':
                y.append('Liability')
            elif i=='avoir des actionnaires':
                y.append('Equity')
            elif i=='hors-bilan':
                y.append('Off-balance sheet')
            elif i=='portefeuille brut':
                y.append('Loan portfolio')
            elif i=='depots us convertis':
                y.append('Deposit in US converted in HTG')
        return y
    df['Parts1_en']=fparts1_en() # write parts1 elements in English
    
###############
    def fparts2_en(): # translation in English
        y=[]
        for i in df.Parts2_fr.tolist():
            if i=='actif':
                y.append('Liquid asset')
            elif i=='disponibilites':
                y.append('Liquid asset')
            elif i=='bons b.r.h.':
                y.append('Non-liquid asset')
            elif i=='depots':
                y.append('Deposit')
            elif i=='obligations a vue':
                y.append('Demand bonds')
            elif i=='obligations a terme':
                y.append('Term bonds')
            elif i=='autres passifs':
                y.append('Other liabilities')
            elif i=='avoir des actionnaires':
                y.append('Equity')
            elif i=='hors-bilan':
                y.append('Off-balance sheet')
            elif i=='portefeuille brut':
                y.append('Loan portfolio')
            elif i=='depots us convertis':
                y.append('Deposit in US converted in HTG')
        return y
    df['Parts2_en']=fparts2_en() # write parts2 elements in English
    
    def fparts2_fr(): # rewrite, with uppercase and accents parts2_fr elements
        y=[]
        for i in df.Parts2_fr.tolist():
            if i=='actif':
                y.append('Actif liquide')
            elif i=='disponibilites':
                y.append('Actif liquide')
            elif i=='bons b.r.h.':
                y.append('Actif illiquide')
            elif i=='depots':
                y.append('Dépôt')
            elif i=='obligations a vue':
                y.append('Obligations à vue')
            elif i=='obligations a terme':
                y.append('Obligations à terme')
            elif i=='autres passifs':
                y.append('Autres passifs')
            elif i=='avoir des actionnaires':
                y.append('Fonds propres')
            elif i=='hors-bilan':
                y.append('Hors-bilan')
            elif i=='portefeuille brut':
                y.append('Portefeuille de crédit')
            elif i=='depots us convertis':
                y.append('Dépôts en USD convertis en HTG')
        return y
    df['Parts2_fr']=fparts2_fr()
    
    def fparts1_fr(): # rewrite, with uppercase and accents parts2_fr elements
        y=[]
        for i in df.Parts1_fr.tolist():
            if i=='actif':
                y.append('Actif')
            elif i=='passif':
                y.append('Passif')
            elif i=='avoir des actionnaires':
                y.append('Fonds propres')
            elif i=='hors-bilan':
                y.append('Hors-bilan')
            elif i=='portefeuille brut':
                y.append('Portefeuille de crédit')
            elif i=='depots us convertis':
                y.append('Dépôts en USD convertis en HTG')
        return y
    df['Parts1_fr']=fparts1_fr()
    
    # remove all rows where there is at least 10 NaN
    df=df.dropna(axis=0,thresh=10)
    
    df=df.reset_index() # reset index
     #########################
    def remove_rowsWithoutIndicatorName(x): # x is a data frames remove row that do not have indicator name - indeed, they are totals
        y_ind=[]
        for i in range(0,len(x.Indicators)):
            if x.Indicators[i]=='nan':
                y_ind.append(False)
            elif x.Indicators[i].strip()=='':
                y_ind.append(False)
             # remove row where indicator titles are: passifs, obligations a vue, obligations a terme (those row are part title and do not have any data.
            elif x.Indicators[i].strip()=='passifs':
                y_ind.append(False)
            elif x.Indicators[i].strip()=='passif':
                y_ind.append(False)
            elif x.Indicators[i].strip()=='depots':
                y_ind.append(False)
            elif x.Indicators[i].strip()=='obligations a vue':
                y_ind.append(False)
            elif x.Indicators[i].strip()=='obligations a terme':
                y_ind.append(False)
            elif x.Indicators[i].strip()=='avoir des actionnaires':
                y_ind.append(False)
                
            else:
                y_ind.append(True)

        return x[y_ind]

    df=remove_rowsWithoutIndicatorName(df)
    df=df.reset_index()
    #######################
    # remove columns of subtotal, since they are several with the same name "SOUS-TOTAL"
    u=[i.startswith('SOUS') for i in df.columns] # identifying the columns where name starts with 'SOUS'
    y=[]  # list of columns names, except those starting with the string 'SOUS'
    for i in range(0,len(u)):
        if u[i]==False:
                   y.append(df.columns[i])
    df=df[y]
    #############################
    # for most of the reports, the total amount of an indicator for the whole banking system is indicating
    # by the column name 'SYSTÈME'. However, for the first reports, this column name is 'TOTAL'. I want to have
    # TOTAL as name for this column for whatever the period considered. It is necessary to have the same name, 
    # since I will have to concatanate all data frames of the balance sheet.
    df.columns=[col.replace('SYSTÈME','TOTAL') for col in df.columns]
    ######################
    # name to identify indicators
    def new_indicatorsNames():
        y=[]
        for i in range(0,len(df.Indicators)):
            y.append(df.Parts2_fr.tolist()[i].replace(' ','_').upper()+'_'+df.Indicators.tolist()[i].replace(' ','_'))
        return y
    df['Indicators']=new_indicatorsNames()
    ##############################
    # add a column indicating the ending month of the report
    def identifyDate(x): # x is the excel data file name
        from datetime import datetime
        Smonth=pd.Series(['3','6','9','12'],index=['2','3','4','1']) # values:months numbers; and index are quarterlies numbers (trim) as indicated by the central bank
        trim=x[5:6]
        if int(trim)==1:
            year=str(int(x[7:11])-1) # the first quarterly of a fiscal year ended in 2021 is the last quarterly in civil year 2020
        else:
            year=x[7:11]
        # identify month of the end of the quarterly:

        date_=datetime.strptime(year+'-'+Smonth[trim]+'-'+'01','%Y-%m-%d') # convert date to datetime format
        return date_
    df['Date']=identifyDate(x)
    return df.iloc[:,2:]
```

```python
def f_combine2(x,fun): # x is the list of files names; fun is the function that import and clean data
    df=pd.DataFrame()
    for file in x:
        df=pd.concat([df,fun(file)])
    
    return df
```
